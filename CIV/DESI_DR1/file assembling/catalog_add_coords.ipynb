```python
from astropy.io import fits
from astropy.table import Table
import numpy as np
```

```python
hdul = fits.open("qso_cat_dr1_main_dark_healpix_zlya-v0.fits") # lya quasar catalog
data = hdul[1].data
hdul = fits.open("valid_QSO_Z.fits") # working catalog
data2 = hdul[1].data

validTID = data2['TARGETID']
lyacatTID = data['TARGETID']
lyaRA = data['TARGET_RA']
lyaDEC = data['TARGET_DEC']

data2 = Table(data2)
data2['RA'] = np.zeros(len(data2))
data2['DEC'] = np.zeros(len(data2))
```

```python
print(data2)
print(data)
```

```python
data2.write('valid_QSO_catalog.fits',"overwrite=True")
```

```python
# Create a dictionary mapping TARGETID to (RA, DEC)
tid_to_coords = {tid: (ra, dec) for tid, ra, dec in zip(lyacatTID, lyaRA, lyaDEC)}

# Loop through each TARGETID in the working catalog
for i, tid in enumerate(validTID):
    if tid in tid_to_coords:
        data2['RA'][i], data2['DEC'][i] = tid_to_coords[tid]
```
