```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
```

```python
hdul = fits.open('39627785479656260.fits')
data=hdul[1].data
wave = data['WAVE']
flux = data['FLUX']
```

```python
plt.plot(wave, flux)
plt.xlabel("Wavelength [Å]")
plt.ylabel("Flux [1e-17 erg/s/cm²/Å]")
plt.grid(True)
plt.legend()
plt.tight_layout()
#plt.xlim(5750,5810)
#plt.ylim(-16,15)
plt.show()
```
